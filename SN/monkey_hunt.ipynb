### ニュートン方程式やモンキーハンティングの説明

ニュートンの運動方程式は以下のように記述されます。
\begin{equation}
  \vec{F}=m\vec{a}
\end{equation}
これは質量mの物体に力$\vec{F}$が働く時に、物体には上式に従う加速度$\vec{a}$が生じるという法則です。
    
**等速直線運動**  
ニュートンの運動方程式において$\vec{F}=0$とすると$\vec{a}=0$となって、物体は加速度を持たなくなってしまいます。加速度がない場合、速度は一定になります。このように物体が常に一定の速度で直進することを等速直線運動といい、このとき、速度$\vec{v}$を持つ物体が時刻$t$だけ変化した時に物体が進んだ距離$\vec{x}$は
\begin{equation}
  \vec{x}=\vec{v}t
\end{equation}
で表されます。

**自由落下運動**  
地表付近にある質量$m$の物体には、地面方向に向かって
\begin{equation}
  F=mg,\quad g=9.8m/s^2
\end{equation}
の力が働きます。これを重力といい、このときの物体が持つ加速度($g$)を重力加速度と言います。
重力加速度$g$を持つ物体の速度は時刻$t$だけ経ったとき$gt$だけ変化します。時刻$t=0$のとき速度$v=0$の物体は時刻$t$のとき速度$v$は$v=gt$と書けて図１のように変化します。  
図1 https://drive.google.com/file/d/186THXTTNMDuDDiTAO1QAQxH6mrWLwRW9/view?usp=share_link  
するとある時刻において微小な時間$Δt$の間に進む距離は青い長方形の面積で与えられ、時刻$t_{0}$までに進んだ距離は黄色の三角形の面積と等しくなり
\begin{equation}
  x(t_{0})=\frac{1}{2}gt_{0}^2
\end{equation}
と求められます。  

**モンキーハンティング**  
銃を持ったハンターが木から落ちたサルを狙ったときに、どのように銃を撃てば弾丸がサルにあたるでしょうか。これがモンキーハンティングと呼ばれる問題です。  
ハンターの位置を原点に取って、ハンターから木までの距離を$l$、木の高さを$h$とします。
時刻$t=0$に猿は自由落下を始め、ハンターは発射角度$θ$、初速$v_{0}$の弾丸を撃ちます。(図２)  
図2 https://drive.google.com/file/d/1jteY8fvR1jZFvBZZMfJlxJABx2H6K37D/view?usp=share_link  
このときの弾丸の運動について考えてみましょう。時刻$t$における弾丸の位置は以下のように記述できます。
\begin{array}{l}
  x_{bullet}=v_{0}\cos \theta⋅t \\
  y_{bullet}=v_{0}\sin \theta⋅t - \frac{1}{2}gt^2
\end{array}
時刻$t$の水平方向の変位$x$は、初速度$v_{0}\cos\theta$の等速直線運動によるもので、鉛直方向の変位$y$は，初速度$v_{0}\sin\theta$の等速直線運動に自由落下による落下距離を加えたものになっていることが分かります。すなわち弾丸の運動は、等速直線運動と自由落下運動を組み合わせた運動とみることができることになる。  
一方、猿の位置は
\begin{array}{l}
  x_{monkey}=h \\
  y_{monkey}=l - \frac{1}{2}gt^2
\end{array}
です。したがって、どちらも時刻$t$における自由落下の距離は等しく、図２のように猿の初期位置に向かって銃を撃てば常に$\vec{AB}∥\vec{PQ}$が成り立ちます。この時の発射角度は$\tan\frac{h}{l}$となっています。つまり、猿から見た弾丸は自身に向かって速さ$v_{0}$の等速直線運動をして近づいてくるように見え、$\vec{PQ}=0$になった時に弾丸が猿にあたります。
では、衝突する時の時刻を$t_{0}$として求めてみると
\begin{equation}
  |\vec{PQ}_{t=t_{0}}|=|\vec{AB}|-v_{0}t_{0}=0 \\
  t_{0}=\frac{|\vec{AB}|}{v_{0}}=\frac{\sqrt{l^2+h^2}}{v_{0}}
\end{equation}
となる。さらに、空中で衝突する条件は$y_{monkey}(t=t_{0})\geqq0$より
\begin{equation}
  y_{monkey}(t=t_{0})=h-\frac{1}{2}gt^2_{0}  \\
  =h-\frac{1}{2}g\frac{l^2+h^2}{v_{0}^2}\geqq0  \\
  ⟹ v_{0}\geqq \sqrt{\frac{g(l^2+h^2)}{2h}}
\end{equation}

### while文

少ない記述量で繰り返し同じ処理をする際に用いられる文法です。2回程度ならまだしも、100回、1000回も同じコードを書くのは大変ですよね。そんなことをせずとも、自分のしたい計算を決められた回数だけ行わせることができるのが繰り返し処理です。

例として以下のコードを見てみましょう。

````python
a = 0
while a < 5:
  print(a)
  a += 1
````

while文ではwhile以下の条件が満たされている限り、その下に書かれた処理を繰り返し行われます。上のコードでは初期値が0の変数aが5より小さい場合にaに1を足し続けている処理が行われています。試しにaが5より小さいという条件を7より小さいという条件に変えてみましょう。出力される数値が増えるはずです。

このように少ない記述量で決められた処理を好きな回数だけ行わせることができます。モンキーハンティング問題でもこの文法は有効です。猿と弾の運動は初期条件と時刻で記述できます。そこで時刻tに0.01秒ずつ加えていき、tが10秒になるまで繰り返し処理で猿と弾の位置を計算するとどうなるでしょうか。実際にコードを書くときに考えてみましょう。

### 関数
プログラミング言語における関数とは変数を受け取り、内部でいくつかの処理を行うものです。(数学的な意味の関数とは異なるので注意してください。) 例えば、上のコードで出てきたprintも関数の一種です。printは１つの変数を受け取り、その結果を画面(厳密には標準出力)に返す関数です。

関数にはprintなどのあらかじめ用意されたものと自分で定義するものがあります。自分で定義する関数の作り方を見てみましょう。以下のコードを見てみてください。

````python
def add(a,b):
  c = a + b
  return c
````

上のように関数は
```
def 関数名(引数1,引数2,...):
  内部の処理
  
  return 戻り値
```
というような形で定義します。ただし、関数名の前には必ず１つ以上の空白、関数名の後ろには空白を空けずに"("、さらに、")"の後ろには":"が必要です。関数名のすぐ後ろのかっこの中には関数に渡す引数を書きます。関数の中ではこの引数の情報を使うことができます。上のaddでは1つ目の引数がa、2つ目の引数がbという名前で定義されています。

def 関数名(引数1,引数2,...):を書いたらすぐ下に内部で行う処理の内容を書いていきます。ただしこのときに処理の内容は字下げ(インデント)して書かなければなりません。例えば,上のaddを
```
def add(a,b):
c = a + b
return c
```
と書くとエラーが出ます。(実際に上のコードをいじって試してみて下さい。)またインデントの文字数は基本的には関数の中で同じでなくてはなりません。例えば,
```
def add(a,b):
  c = a + b
   return c
```
と書くとエラーが出ます。このように関数を定義するときにはインデントの文字数がそろっているか注意してください。

関数の処理の内容に話を戻します。関数の中ではaとbの値の和をcに代入しています。そして関数の内部の処理の最後でreturn 戻り値(上のaddではc)という文を書いて関数の値を出力しています。

次にこの関数を実際に使ってみましょう。以下のコードを実行してみて下さい。

````python
a = add(1,1)
print(a)
````

それでは上のposition関数を使って弾とサルの運動を計算してみましょう。以下の図のような状況を考えます。(後で図を足す)

````python
def calc_all_position(t, vx_bullet, vy_bullet, l_monkey, h_monkey):
    x_bullet, y_bullet = position(t, 0, 0, vx_bullet, vy_bullet)
    x_monkey, y_monkey = position(t, l_monkey, h_monkey, 0, 0)
    return x_bullet, y_bullet, x_monkey, y_monkey
````

````python
import matplotlib.pyplot as plt
from matplotlib import patches
import numpy as np

vx_bullet = 20 #弾の初速のx成分
vy_bullet = 20 #弾の初速のy成分
l_monkey = 20 #猿と弾の間の水平距離
h_monkey = 20
r_monkey = 0.5

#ここから下は変更しないでください。
t_list = []
x_bullet_list = []
y_bullet_list = []
x_monkey_list = []
y_monkey_list = []
fig, ax = plt.subplots()
bullet_color = "blue"
monkey_color = "orange"

for t in np.arange(0,10,0.01):
  x_bullet, y_bullet, x_monkey, y_monkey = calc_all_position(t, vx_bullet, vy_bullet, l_monkey, h_monkey)
  x_bullet_list.append(x_bullet)
  y_bullet_list.append(y_bullet)
  x_monkey_list.append(x_monkey)
  y_monkey_list.append(y_monkey)
  circle_monkey = patches.Circle(xy=(x_monkey, y_monkey),
                                 radius=r_monkey,
                                 color=monkey_color, alpha=0.5)
  ax.add_patch(circle_monkey)

ax.plot(x_bullet_list, y_bullet_list,label = "bullet")
ax.plot(x_monkey_list, y_monkey_list,label = monkey_color)
ax.legend()
plt.show()
````

弾(bullet)とサル(monkey)が無限に下まで落ちていってしまいます。そこで弾とサルのどちらかが地面に着いた瞬間に時間を止めてしまいましょう。

````python
def calc_all_position(t, t_land, vx_bullet, vy_bullet, l_monkey, h_monkey, r_monkey):
    if t_land >= 0:
      t = t_land
    x_bullet, y_bullet = position(t, 0, 0, vx_bullet, vy_bullet)
    x_monkey, y_monkey = position(t, l_monkey, h_monkey, 0, 0)
    if (y_bullet < 0) or (y_monkey < r_monkey):
      t_land = t
    return t_land, x_bullet, y_bullet, x_monkey, y_monkey
````

上のコードでは初めにt_landが正だった場合、時刻tをt_landに設定するようにしています。また、最後に弾のy座標が地面(y=0)より小さくなった場合にその時の時刻tをt_landに代入しています。

### 問題
